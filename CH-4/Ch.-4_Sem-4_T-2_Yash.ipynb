### Feature Transformation :

```python
# Feature Additon
import pandas as pd
room_length = [18,20,10,12,18,11]
room_breadth = [20,20,10,11,19,10]
room_type = ["Big","Big","Normal","Normal","Big","Normal"]
```

```python
data = pd.DataFrame({
    "length":room_length,
    "Breadth":room_breadth,
    "Type":room_type
})
data
```

```python
data["Area"] = data["length"]*data["Breadth"]
data
# EDA : Exploratory Data Analysis
```

```python
pd.get_dummies(data) # categorical data ne  column ma convert kare.
```

### drop_first = True : - first column remove kare

```python
data = pd.get_dummies(data,drop_first=True)
data
```
